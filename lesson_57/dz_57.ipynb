```python
import matplotlib.pyplot as plt
from PIL import Image
```

# Функція зчитування та відображення зображення

```python
def show_image(file_path):
   img = Image.open(file_path)
   plt.imshow(img)
   plt.title('Оригінальне фото')
   plt.show()
```

# Функція відображення до і після коригування

```python
def show_compare(original_image, modif_image, modif_name):
    # Виводимо зображення до і після коригування 
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(original_image)
    axes[0].set_title('Oригінальне фото')
    axes[1].imshow(modif_image)
    axes[1].set_title(f'{modif_name} фото')
    plt.show()
```

# Збереження зображення на диск

```python
def save_image(image, output_path):
       image.save(output_path)
```

# Функція застосування коригування контрастності та яскравості

```python
def adjust_contrast_brightness(image, contrast_factor, brightness_factor):
    from PIL import ImageEnhance
    enhancer = ImageEnhance.Brightness(image)
    img_brightness = enhancer.enhance(brightness_factor)
    enhancer = ImageEnhance.Contrast(img_brightness)
    img_contrast = enhancer.enhance(contrast_factor)
    return img_contrast
```

# Функція застосування коригування яскравості

```python
def adjust_brightness(image, brightness_factor):
    from PIL import ImageEnhance
    enhancer = ImageEnhance.Brightness(image)
    img_brightness = enhancer.enhance(brightness_factor)
    return img_brightness
```

# Функція застосування коригування насиченості

```python
def adjust_saturation(image, saturation_factor):
    from PIL import ImageEnhance
    enhancer = ImageEnhance.Color(image)
    img_saturation = enhancer.enhance(saturation_factor)
    return img_saturation
```

# Функція застосування зміни розміру зображення

```python
def resize_image(image, new_width, new_height):
    resized_image = image.resize((new_width, new_height))
    return resized_image
```

# Функція застосування обрізання зображення

```python
def crop_image(image, left, top, right, bottom):
    cropped_image = image.crop((left, top, right, bottom))
    return cropped_image
```

# Функція застосування повороту зображення

```python
def rotate_image(image, angle):
    rotated_image = image.rotate(angle)
    return rotated_image
```

# Приклад використання всіх функцій

```python
file_path = 'img_correct_03.jpg'
img = Image.open(file_path)
show_image(file_path)
```

```python
# Зміна контрастності і яскравості
adjusted_image = adjust_contrast_brightness(img, 1.5, 1.2)
show_compare(img, adjusted_image, 'Скориговані контраст і яскравість')
```

```python
# Зміна яскравості
brightness_adjusted_image = adjust_brightness(img, 1.4)
show_compare(img, brightness_adjusted_image, 'Скоригована яскравість')
```

```python
# Зміна насиченості
saturation_adjusted_image = adjust_saturation(img, 1.5)
show_compare(img, saturation_adjusted_image, 'Скоригована насиченість')
```

```python
# Зміна розміру зображення
resized_image = resize_image(img, 300, 200)
show_compare(img, resized_image, 'Змінено розмір')
```

```python
# Обрізання зображення
cropped_image = crop_image(img, 50, 50, 200, 200)
show_compare(img, cropped_image, 'Обрізане')
```

```python
# Поворот зображення
rotated_image = rotate_image(img, 45)
show_compare(img, rotated_image, 'Повернуте під кутом')
```

```python
# Збереження зміненого зображення 
save_image(cropped_image, 'cropped_image.jpg')
```
