```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Функції завантаження та відображення зображення

```python
def load_image(image_path):
    # Завантаження зображення з файлу
    image = cv2.imread(image_path)
    return image
def show_image(image, title='', cmap=None):
    # Відображення зображення
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

#### Приклад застосуваня функції 

```python
image_path = 'cat.jpg'
original_image = load_image(image_path)
show_image(original_image, title='Оригінальне зображення')
```

## Функція застосування фільтра Собеля

```python
def apply_sobel(image):
    # Застосовуємо фільтр Собеля
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    return gradient_magnitude
```

#### Приклад застосуваня фільтра Собеля

```python
sobel_result = apply_sobel(original_image)
show_image(sobel_result, title='Фільтр Собеля', cmap='gray')
```

## Функція застосування медіанного фільтра

```python
def apply_median_filter(image, kernel_size=3):
    # Перетворюємо зображення у відтінки сірого
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Застосовуємо медіанний фільтр
    median_filtered = cv2.medianBlur(gray_image, kernel_size)
    # Розширюємо розмірність до 3 каналів для правильного відображення
    median_filtered_colored = cv2.cvtColor(median_filtered, cv2.COLOR_GRAY2BGR)
    return median_filtered_colored
```

#### Приклад застосуваня медіанного фільтра

```python
kernel_size = 5  
median_filtered_result = apply_median_filter(original_image, kernel_size)
show_image(median_filtered_result, title='Медіанний фільтр')
```
