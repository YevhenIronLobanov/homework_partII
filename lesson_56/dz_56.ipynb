```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = cv2.imread ('cat.jpg')
type(im)
```

```python
plt.imshow(im)
```

# Конвертація в RGB

```python
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(im_rgb)
```

# Декомпозиція по складових RGB моделі

```python
R = im_rgb.copy() # створення копії
```

```python
# По червоному кольору
R[:, :, (1,2)]=0
```

```python
plt.imshow(R)
```

```python
# По зеленому кольору
G = im_rgb.copy()
G[:, :, (0,2)]=0
plt.imshow(G)
```

```python
# По блакитному кольору
B = im_rgb.copy()
B[:, :, (0,1)]=0
plt.imshow(B)
```

# Конвертація в HSV

```python
im_hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
```

# Декомпозиція по складових HSV моделі

```python
# Копіюємо зображення у просторі кольорів HSV
H = im_hsv.copy()
```

```python
# Встановлюємо значення для  насиченості (S) та значення (V)
# Значення відтінку (H) залишаємо незмінним (максимальне - 179)
H[:, :, (1, 2)] = 225 
```

```python
# Для візуалізації конвертуємо до RGB
H_rgb = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.imshow(H_rgb)
```

```python
# Копіюємо зображення у просторі кольорів HSV
S = im_hsv.copy()
```

```python
# Встановлюємо значення для  відтінку (H) та значення (V)
# Значення насиченності (S) залишаємо незмінним (максимальне - 255)
S[:, :, 0] = 150
S[:, :, 2] =225
```

```python
# Для візуалізації конвертуємо до RGB
S_rgb = cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.imshow(S_rgb)
```

```python
# Копіюємо зображення у просторі кольорів HSVV
V = im_hsv.copy()
```

```python
# Встановлюємо значення для  відтінку (H) та насиченності (S)
# Значення (V) залишаємо незмінним (максимальне - 255)
V[:, :, 0] = 150
V[:, :, 1] =225
```

```python
# Для візуалізації конвертуємо до RGB
V_rgb = cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.imshow(V_rgb)
```

# Колірна модель CIE LAB

```python
# Завапнтаження необхідної бібліотеки
import matplotlib.colors as clr
```

```python
# Конвертація в CIE_lAB
```

```python
im_CL = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
```

# Декомпозиція по складових CIE LAB моделі

```python
L = im_CL.copy()
```

```python
# Встановлюємо нульові значення для A (червоний-зелений) та B (синій-жовтий)
L[:, :, (1, 2)] = 0
```

```python
# Для візуалізації конвертуємо до RGB та конвертуємо зображення RGB у відтінки сірого
L_rgb = cv2.cvtColor(L, cv2.COLOR_LAB2RGB)
L_gray = cv2.cvtColor(L_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(L_gray, cmap='gray')
```

```python
A = im_CL.copy()
```

```python
# Встановлюємо значення для L (яскравість) та для B (синій-жовтий)
A [:, :, 0] = 127
A [:, :, 2] = 0
```

```python
# Для візуалізації конвертуємо до RGB та конвертуємо зображення RGB у відтінки сірого
A_RGB = cv2.cvtColor(A, cv2.COLOR_LAB2RGB)
A_GRAY = cv2.cvtColor(A_RGB, cv2.COLOR_RGB2GRAY)
```

```python
# Створення кольорової карти з трьома кольорами (Red, Gray, Green) та 255 відтінками (N=255)
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red', 'Gray', 'Green'], N=255)
plt.imshow(A_GRAY, cmap=cmap_a)
```

```python
B = im_CL.copy()
```

```python
# Встановлюємо значення для L (яскравість) та для A (червоний-зелений)
B [:, :, 0] = 127
B [:, :, 1] = 0
```

```python
# Для візуалізації конвертуємо до RGB та конвертуємо зображення RGB у відтінки сірого
B_RGB = cv2.cvtColor(B, cv2.COLOR_LAB2RGB)
B_GRAY = cv2.cvtColor(B_RGB, cv2.COLOR_RGB2GRAY)
```

```python
# Створення кольорової карти з трьома кольорами (Yellow, Gray, Blue) та 255 відтінками (N=255)
cmap_b = clr.LinearSegmentedColormap.from_list('custom blue', ['Yellow', 'Gray', 'Blue'], N=255)
plt.imshow(B_GRAY, cmap=cmap_b)
```
